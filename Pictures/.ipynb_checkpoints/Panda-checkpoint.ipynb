```python
import pygame
import numpy as np
from pygame.draw import *


pygame.init()

FPS = 30
screen = pygame.display.set_mode((1700, 1000))

    
arc(screen,'black',[20,220,250,200], 0, np.pi/2, 10)


def background():
    rect(screen, (255,175,128), (0, 0, 1700, 1000))

    
def back_left_leg():
    surface_back_left_leg = pygame.Surface((1700, 1000), pygame.SRCALPHA)
    surface_back_left_leg.fill ((255,255,255,0))
    ellipse(surface_back_left_leg, 'black', (1200, 150, 100,200))
    polygon(surface_back_left_leg, 'black' , [[1210, 280], [1200, 350], [1310, 350], [1300, 260]])
    surface_back_left_leg = pygame.transform.rotate(surface_back_left_leg, -25)
    screen.blit(surface_back_left_leg, (0, 0))
    surface_back_left_leg_1 = pygame.Surface((1700, 1000), pygame.SRCALPHA)
    surface_back_left_leg_1.fill ((255,255,255,0))
    ellipse(surface_back_left_leg_1, 'black', (1125, 185, 100,50))
    surface_back_left_leg_1 = pygame.transform.rotate(surface_back_left_leg_1, -35)
    screen.blit(surface_back_left_leg_1, (0, 0))
    
    
    
def body():
    surface_body = pygame.Surface((1700, 1000), pygame.SRCALPHA)
    surface_body.fill ((255,255,255,0))
    ellipse(surface_body, 'white', (1100, 330, 350,200))
    surface_body = pygame.transform.rotate(surface_body, -10)
    screen.blit(surface_body, (0, 0))
    
def forward_left_leg():
    polygon(screen, 'black' , [[1380, 540], [1370, 560], [1300, 850], [1250, 900], [1300, 950], [1380, 900], [1380, 850],[1420,700]])
    surface_forward_left_leg = pygame.Surface((1700, 1000), pygame.SRCALPHA)    
    surface_forward_left_leg.fill ((255,255,255,0))
    ellipse(surface_forward_left_leg, 'black', (980, 130, 75,50))
    surface_forward_left_leg = pygame.transform.rotate(surface_forward_left_leg, -55)
    screen.blit(surface_forward_left_leg, (0, 0))
    surface_forward_left_leg_1 = pygame.Surface((1700, 1000), pygame.SRCALPHA)    
    surface_forward_left_leg_1.fill ((255,255,255,0))
    ellipse(surface_forward_left_leg_1, 'black', (1080, 650, 110,80))
    surface_forward_left_leg_1 = pygame.transform.rotate(surface_forward_left_leg_1, 45)
    screen.blit(surface_forward_left_leg_1, (0, 0))
    
 
def head ():
    circle(screen, "white", (1200,550), 130)
    ellipse(screen, 'white', (1070, 440, 100,210))
    ellipse(screen, 'white', (1090, 380, 200,100))
    circle(screen, "white", (1105,450), 30)
    circle(screen, "white", (1105,450), 30)
    circle(screen, "white", (1115,618), 30)
    circle(screen, "white", (1275,455), 40)
    circle(screen, "black", (1170,580), 30)
    ellipse(screen, 'black', (1070, 540, 50,70))
    ellipse(screen, 'black', (1070, 620, 70,40))
    surface_head = pygame.Surface((1700, 1000), pygame.SRCALPHA)    
    surface_head.fill ((255,255,255,0))
    ellipse(surface_head, 'black', (1300,  220, 110,50))
    surface_head = pygame.transform.rotate(surface_head, 45)
    screen.blit(surface_head, (0, 0))    
    surface_head_1 = pygame.Surface((1700, 1000), pygame.SRCALPHA)    
    surface_head_1.fill ((255,255,255,0))
    ellipse(surface_head_1, 'black', (810,  -10, 100,60))
    surface_head_1 = pygame.transform.rotate(surface_head_1, -30)
    screen.blit(surface_head_1, (0, 0))    
    
def forward_right_leg():    
    ellipse(screen, 'black', (1170, 640, 100,200))
    
    
background()
body()
back_left_leg()    
forward_left_leg()
forward_right_leg()

head()


pygame.display.update()
clock = pygame.time.Clock()
finished = False

while not finished:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```
